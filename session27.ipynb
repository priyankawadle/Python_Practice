```python
unique_stores =sales_data.groupby("Region")["Store ID"].unique().reset_index()
unique_stores.columns=["Regions", "Unique Store IDs"]
unique_stores
```

```python
sales_data["Date"] = pd.to_datetime(sales_data["Date"])
sales_data["Day_Name"] = sales_data["Date"].dt.day_name()
sales_data["Month_Name"] = sales_data["Date"].dt.month_name()
sales_data
```

```python
# sales_data.groupby("Day_Name")["Product ID"].value_counts().reset_index()
avg_sale_per_day_name =sales_data.groupby(["Day_Name","Product ID"]).agg({"Units Sold": "mean"}).reset_index()
ids =avg_sale_per_day_name.groupby(["Day_Name"])["Units Sold"].idxmax()
ids
```
