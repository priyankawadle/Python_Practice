```python
#

#Joins : Combining two datasets

import pandas as pd


order_details={"CustomerID":['C1','C2','C3','C4'],
          "CustomerNames":['Atul','Jit','Pratyuksha','Hardik'],
          "Age":[23,24,25,26],
          "City":['Delhi','Mumbai','Delhi','Mumbai'],
          "NumOfOrders":[10,20,30,40]}
order_details_df=pd.DataFrame(order_details)

return_details={"CustomerID":['C1','C2'],
          "NumReturns":[5,6]}

return_details_df=pd.DataFrame(return_details)




```

```python
return_order_Details = order_details_df.merge(return_details_df,how='left',on='CustomerID')
```

```python
#Product Numbers:
# What is the average return percentage for orders from Delhi and Mumbai ?
return_order_Details['PercentageOfReturns']=round((return_order_Details['NumReturns']*100)/return_order_Details['NumOfOrders'],2)
return_order_Details
```

```python
return_order_Details.groupby("City")["PercentageOfReturns"].mean()
```

Given Tables:

- Customers(CustomerID, Name, City)
- Orders(OrderID, CustomerID, RestaurantID, Amount)
- Ratings(CustomerID, RestaurantID, Rating)

1. Join the tables appropriately to:
- Count the number of orders per customer.
- Identify customers who have rated any restaurant ≥ 4.
2. Filter the results to return only:
- Customers who have placed more than 2 orders, and
- Given at least one rating ≥ 4.

```python
import pandas as pd

customers = pd.DataFrame({
    'CustomerID': ['C1', 'C2', 'C3', 'C4'],
    'Name': ['Alice', 'Bob', 'Charlie', 'David'],
    'City': ['Mumbai', 'Delhi', 'Bangalore', 'Pune']
})

orders = pd.DataFrame({
    'OrderID': ['O1', 'O2', 'O3', 'O4', 'O5', 'O6', 'O7'],
    'CustomerID': ['C1', 'C1', 'C1', 'C2', 'C3', 'C3', 'C3'],
    'RestaurantID': ['R1', 'R2', 'R1', 'R2', 'R1', 'R2', 'R3'],
    'Amount': [250, 150, 200, 300, 180, 220, 210]
})

ratings = pd.DataFrame({
    'CustomerID': ['C1', 'C2', 'C3', 'C3', 'C4'],
    'RestaurantID': ['R1', 'R2', 'R1', 'R3', 'R2'],
    'Rating': [5, 3, 4, 2, 5]
})

# Preview the data
print("Customers:\n", customers, "\n")
print("Orders:\n", orders, "\n")
print("Ratings:\n", ratings)
```

```python
# - Count the number of orders per customer.
customer_order =customers.merge(orders,how="left",on="CustomerID")
customer_order_count =customer_order.groupby("CustomerID")["OrderID"].count().reset_index()
customer_order_count
```

```python
# - Identify customers who have rated any restaurant ≥ 4.
customer_rating =customers.merge(ratings,how="left",on="CustomerID")
customer_rating_greater =customer_rating[customer_rating["Rating"]>=4]
customer_rating_greater
```

```python
# Customers who have placed more than 2 orders, and Given at least one rating ≥ 4.
customer_order_count_rating =customer_order_count.merge(customer_rating_greater,how="inner",on="CustomerID")
customer_order_count_rating[customer_order_count_rating["OrderID"] > 2]
```
